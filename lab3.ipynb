```python
#Câu 1
print('Câu 1')
from scipy import linalg # Tải gói linalg của scipy vào bộ nhớ (để sử dụng)
import numpy as np # Tải gói numpy vào sử dụng với tên gọi là np
a = np.array([1, 2, 3])
b = np.array([(1+9j, 2j, 3j), (4j, 5j, 6j)]) # số phức
c = np.array([[ (0.5, 1.5, 10), (3,2,1) ] , [(6,5,4), (7,8,9)]]) # ma trận số thực
A = np.matrix(np.random.random( (2,2) ) )
B = np.asmatrix(b) # Chuyển b thành dạng ma trận
C = np.asmatrix(np.random.random( (10,5) ) )
D = np.asmatrix([ [4, 3], [2, 6] ])
F = np.eye(3, k=1) # tạo ma trận đường chéo. 3 là ma trận 3x3, k=1 là đường chéo nằm phía trên đường chéo chính (k = 0).
print(f'Kết quả của 3, k=1 là : {F}')
F1 = np.eye(3, k=0)
print(f'Kết quả của 3, k=0 là : {F1}')
F2 = np.eye(3, k=-1)
print(f'Kết quả của 3, k=-1 là : {F2}')
print('Xem hạng ma trận:')
Hang = np.linalg.matrix_rank(C)
print(Hang)
print('Tính ma trận nghịch đảo:')
print(A.I)
linalg.inv(A) # đây là lệnh của gói scipy, nên có thể sử dụng scipy.linalg.inv(A)
print('Định thức ma trận:')
dinhthuc = linalg.det(A)
print(dinhthuc)
print('Chuyển đổi ma trận:')
dong_cot = A.T
print(f'dòng thành cột: {dong_cot}')
vi_lien_hop = A.H
print(f'Chuyển vị liên hợp: {vi_lien_hop}')
print('Giải các loại phương trình tuyến tính:')
pttt_1 = linalg.solve(A, b)
print(pttt_1)
E = np.asmatrix(a).T
print(E)
pttt_2 = linalg.lstsq(F, E)
print(pttt_2)
print('Các hàm trên ma trận:')
print('Cộng ma trận:')
cong = np.add(A,D)
print(cong)
print('Trừ ma trận:')
tru = np.subtract(A,D)
print(tru)
print('Chia ma trận:')
chia = np.divide(A,D)
print(chia)
print('Nhân ma trận:')
nhan = np.dot(A,D)
nhan1 = A@D
nhan2 = np.multiply(D,A)
nhan3 = np.vdot(A,D)
print(nhan)
print(nhan1)
print(nhan2)
print(nhan3)
print('Hàm lũy thừa và logarith ma trận:')
luythua = linalg.expm(A)
print(luythua)
logarith = linalg.logm(A)
print(logarith)
```

```python
print('Câu 4.1')
import numpy as np
A = np.array([17, 34, 5])
B = np.array([[1, -2, 2], [-1, 1, 3], [1, -1, -4]])
AB = np.dot(A,B)
print("Kết quả 3:")
print(AB)

C = np.array([0, 6, 26])
CB = np.dot(C,B)
print('Kết quả 4:')
print(CB)

E = np.array([33, 0, 14])
EB = np.dot(E,B)
print('Kết quả 5:')
print(EB)

G = np.array([12, 9, 34])
GB = np.dot(G,B)
print('Kết quả 6:')
print(GB)

D = np.array([25, 35, 35])
DB = np.dot(D,B)
print('Kết quả 7:')
print(DB)

inv_B = np.linalg.inv(B)
print(f'Ma trận nghịch đảo của B là: \n {inv_B}')
print('Phục hồi các kết quả là: ')
kq3 = np.dot(AB,inv_B)
print(kq3)
kq4 = np.dot(CB,inv_B)
print(kq4)
kq5 = np.dot(EB,inv_B)
print(kq5)
kq6 = np.dot(GB,inv_B)
print(kq6)
kq7 = np.dot(DB,inv_B)
print(kq7)
```

```python
print('Câu 2')
print('a. ')
import numpy as np
n = int(input("Nhập số lượng phần tử c: "))
c = []
for i in range(n):
    x = int(input(f"c[{i}] = "))
    c.append(x)
# Ma trận đơn vị
M = np.eye(2, dtype=int)
# Nhân các ma trận
for x in c:
    A = np.array([
        [x, 1],
        [1, 0]
    ])
    M = np.dot(M, A)
# Lấy pn và qn
pn = M[0, 0]
qn = M[1, 0]
print("\nMa trận kết quả:")
print(M)
print("pn =", pn)
print("qn =", qn)
print(f"\nPhân số thu được: {pn}/{qn}")
```
